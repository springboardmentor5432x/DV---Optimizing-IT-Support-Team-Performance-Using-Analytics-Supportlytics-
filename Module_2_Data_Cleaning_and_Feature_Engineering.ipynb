```python
import pandas as pd

df = pd.read_csv("Technical Support Dataset 3.csv")
```

Check Missing Values

```python
df.isnull().sum()
```

```python
df['Created time'] = pd.to_datetime(df['Created time'], errors='coerce')
df['Resolution time'] = pd.to_datetime(df['Resolution time'], errors='coerce')
```

Create Resolution_Duration

```python
df['Resolution_Duration'] = ( df['Resolution time'] - df['Created time']).dt.total_seconds() / 3600
```

```python
df['Resolution_Duration'] = df['Resolution_Duration'].abs()
```

Create Priority_Score

```python
priority_map = {
    'High': 3,
    'Medium': 2,
    'Low': 1,
}
df['Priority_Score'] = df['Priority'].map(priority_map)
```

Save the Processed Dataset

```python
df.to_csv("technical_support_cleaned.csv", index=False)
```
