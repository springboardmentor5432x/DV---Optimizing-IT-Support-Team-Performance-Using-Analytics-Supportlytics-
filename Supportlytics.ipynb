```python
import pandas as pd
import numpy as np
```

```python
file_path = 'Technical Support Dataset.csv'
df = pd.read_csv('/content/Technical Support Dataset.csv')
```

```python
print(df.isnull().sum())
```

```python
print("\nDistribution by Product Group:")
print(df['Product group'].value_counts())

```

```python
print("\nDistribution by Priority:")
print(df['Priority'].value_counts())

```

```python
print("\nDistribution by Topic:")
print(df['Topic'].value_counts())
```

```python
df['Topic'] = df['Topic'].str.strip().str.capitalize()
```

```python
date_columns = ['Created time', 'First response time', 'Resolution time', 'Close time']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df['Resolution_Duration'] = (df['Resolution time'] - df['Created time']).dt.total_seconds() / 3600
```

```python
df['Response_Duration'] = (df['First response time'] - df['Created time']).dt.total_seconds() / 3600
```

```python
priority_map = {'Low': 1, 'Medium': 2, 'High': 3}
df['Priority_Score'] = df['Priority'].map(priority_map)
```

```python
df['Performance_Bucket'] = pd.qcut(df['Resolution_Duration'].dropna(), q=3, labels=['Fast', 'Average', 'Slow'])
```

```python
df['Agent interactions'] = df['Agent interactions'].fillna(0)
df.to_csv('/content/Technical Support Dataset.csv', index=False)
print(df[['Ticket ID', 'Resolution_Duration', 'Priority_Score', 'Performance_Bucket']].head())
```
