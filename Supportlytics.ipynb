```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 1. Load the Dataset
# Make sure you have uploaded your CSV to the Colab 'Files' folder on the left!
df = pd.read_csv('/content/dataset-tickets-multi-lang-4-20k.csv')

# --- FIX: Check the actual column names ---
print("Your current columns are:", df.columns.tolist())

# If your ID column is named something else, change 'Ticket_ID' below
# to match what you see in the list above.
id_column = df.columns[0] # This automatically picks the first column as the ID

# 2. Synthetic Feature Engineering: Generating Dates
np.random.seed(42)
start_date = datetime(2025, 1, 1)

# Create 'Ticket_Creation_Date'
df['Ticket_Creation_Date'] = [start_date + timedelta(days=np.random.randint(0, 180)) for _ in range(len(df))]

# Create 'Resolution_Date' (Creation Date + 1 to 7 days)
df['Resolution_Time_Days'] = np.random.randint(1, 8, size=len(df))
df['Resolution_Date'] = df['Ticket_Creation_Date'] + pd.to_timedelta(df['Resolution_Time_Days'], unit='d')

# 3. Initial Data Quality Check
print("\nFirst 5 Rows of Engineered Data:")
# Using id_column variable so it doesn't crash again!
print(df[[id_column, 'Ticket_Creation_Date', 'Resolution_Date', 'Resolution_Time_Days']].head())

# 4. Save this version
df.to_csv('Supportlytics_Phase1_Data.csv', index=False)
```
