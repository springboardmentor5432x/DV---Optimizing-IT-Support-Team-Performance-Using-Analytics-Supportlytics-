```python
import pandas as pd

df = pd.read_csv("customer_support_tickets.csv")
original_shape = df.shape
original_shape
```

```python
df.head()
```

```python
df.drop(columns=[
    "Customer Name",
    "Customer Email",
    "Ticket Channel",
    "Customer Satisfaction Rating",
    "Resolution"
], inplace=True)
```

```python
df.head()
```

```python
priority_map = {
    "Critical": 4,
    "High": 3,
    "Medium": 2,
    "Low": 1
}

df['Priority Score'] = df['Ticket Priority'].map(priority_map)
```

```python
df.head()
```

```python
df['Ticket Priority'].value_counts(dropna=False)
```

```python
df['First Response Time'] = pd.to_datetime(
    df['First Response Time'], 
    format="%d-%m-%Y %H:%M", 
    errors='coerce'
)
```

```python
df['Time to Resolution'] = pd.to_datetime(
    df['Time to Resolution'], 
    format="%d-%m-%Y %H:%M", 
    errors='coerce'
)
```

```python
df['First Response Time'].head()
df['Time to Resolution'].head()
```

```python
df.head()
```

```python
print(df[['Date of Purchase','First Response Time','Time to Resolution']].head())
```

```python
df['Date of Purchase'] = pd.to_datetime(
    df['Date of Purchase'],
    format="%d-%m-%Y",
    errors='coerce'
)
```

```python
print(df[['Date of Purchase','First Response Time','Time to Resolution']].head())
```

```python
# STEP 2 — Compute Response Delay, resultion time, total handling in hours
```

```python
df['Response Delay (Hours)'] = (
    df['First Response Time'] - df['Date of Purchase']
).dt.total_seconds() / 3600
```

```python
df['Resolution Time (Hours)'] = (
    df['Time to Resolution'] - df['First Response Time']
).dt.total_seconds() / 3600
```

```python
df['Total Handling (Hours)'] = (
    df['Time to Resolution'] - df['Date of Purchase']
).dt.total_seconds() / 3600
```

```python
#If Time to Resolution is NaT → unresolved
df['Unresolved'] = df['Time to Resolution'].isna().astype(int)
```

```python
#If First Response Time is NaT → unresponded

df['Unresponded'] = df['First Response Time'].isna().astype(int)
```

```python
#KPI categorization
```

```python
df['Resolution Category'] = pd.cut(
    df['Resolution Time (Hours)'],
    bins=[0, 24, 72, float('inf')],
    labels=['Fast', 'Medium', 'Slow']
)
```

```python
df[['Response Delay (Hours)', 'Resolution Time (Hours)', 'Total Handling (Hours)', 'Unresolved', 'Unresponded']].head()
```

```python
df.drop(columns=[
    'Response Delay (Hours)',
    'Total Handling (Hours)'
], inplace=True)
```

```python
df.loc[df['Resolution Time (Hours)'] < 0, 'Resolution Time (Hours)'] += 24
```
