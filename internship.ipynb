```python
#Handle Missing Values
df = df.dropna(subset=["Ticket Priority", "Ticket Status"])
```

```python
#Convert Time Columns to Numeric
df["First Response Time"] = pd.to_numeric(df["First Response Time"], errors="coerce")
df["Time to Resolution"] = pd.to_numeric(df["Time to Resolution"], errors="coerce")
```

For further analysis columns are converted to numeric format like float or int from object

```python
# Convert Priority into Numbers
priority_map = {
    "Low": 1,
    "Medium": 2,
    "High": 3,
    "Critical": 4
}

df["Priority_Score"] = df["Ticket Priority"].map(priority_map)

```

```python
df
```

Made new column named Priority_Score in which these categories are given numeric values

```python
#Group Customer Satisfaction Levels
df["Satisfaction_Level"] = pd.cut(
    df["Customer Satisfaction Rating"],
    bins=[0, 2, 4, 5],
    labels=["Low", "Medium", "High"]
)
```

Converts numeric ratings (15) into simple groups:

12 - Low satisfaction

34 - Medium satisfaction

5 - High satisfaction

and creates a new column Satisfaction_Level

```python
#Save the Cleaned Dataset
df.to_csv("supportlytics_cleaned_dataset.csv", index=False)
```
