```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_it_support_tickets.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
df['Ticket Type'].value_counts().plot(kind='bar')
plt.title("Ticket Type Distribution")
plt.xlabel("Ticket Type")
plt.ylabel("Number of Tickets")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Ticket Priority', data=df)
plt.title("Ticket Priority Distribution")
plt.xlabel("Ticket Priority")
plt.ylabel("Number of Tickets")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['Ticket Status'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Ticket Status Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='Ticket Channel', data=df)
plt.title("Ticket Channel Distribution")
plt.xlabel("Number of Tickets")
plt.ylabel("Ticket Channel")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Ticket Priority', y='Priority_Score', data=df)
plt.title("Ticket Priority vs Priority Score")
plt.xlabel("Ticket Priority")
plt.ylabel("Priority Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Customer Satisfaction Rating'], bins=5)
plt.title("Customer Satisfaction Rating Distribution")
plt.xlabel("Satisfaction Rating")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['Ticket Subject'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Ticket Subjects")
plt.xlabel("Ticket Subject")
plt.ylabel("Number of Tickets")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Priority_Score',
    y='Customer Satisfaction Rating',
    data=df
)
plt.title("Priority Score vs Customer Satisfaction")
plt.xlabel("Priority Score")
plt.ylabel("Customer Satisfaction Rating")
plt.show()
```
